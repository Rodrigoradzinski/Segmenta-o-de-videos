```python
"""
O  Codigo consiste em uma série de funções para realizar a segmentação de imagens, 
converter rótulos de segmentação em imagens coloridas e fornecer mapas de cores e nomes de classes pré-definidos para visualização e interpretação dos resultados da segmentação.
"""
def label_to_color_image(label):
      # retorna erro se o rótulo de entrada não possui 2 dimensões
    if label.ndim != 2:
        raise ValueError('Expect 2-D input label')
    colormap = criar_colormap()
    # retorna erro se o valor/índice do rótulo é maior que o valor de indice maximo permitida para a lista de cores.
    if np.max(label) >= len(colormap):
        raise ValueError('label value too large.')

    return colormap[label]
```

```python
"""
A classe ModeloDeepLab carrega um modelo pré-treinado DeepLab de um arquivo 
e utiliza esse modelo para fazer a inferência de segmentação semântica em imagens. 
Em termos simples, a classe utiliza o modelo para categorizar cada pixel de uma imagem de acordo com o objeto que o pixel representa.
"""
class ModeloDeepLab(object):
    """Classe para carregar o modelo deeplab e fazer a inferência."""

    FROZEN_GRAPH_NAME = 'frozen_inference_graph'

    def __init__(self, tarball_path):
        """ Cria e carrega o model deeplab pré-treinado. """
        self.graph = tf.Graph()
        graph_def = None

        # Extrai os frozen graph do arquivo tar.
        tar_file = tarfile.open(tarball_path)
        for tar_info in tar_file.getmembers():
            if self.FROZEN_GRAPH_NAME in os.path.basename(tar_info.name):
                file_handle = tar_file.extractfile(tar_info)
                graph_def = tf.compat.v1.GraphDef.FromString(file_handle.read())
                break
        tar_file.close()

        if graph_def is None:
            raise RuntimeError('Nao foi possivel encontrar o inference graph no arquivo tar.')

        with self.graph.as_default():
            tf.import_graph_def(graph_def, name='')

        self.sess = tf.compat.v1.Session(graph=self.graph)

    def run(self, image, INPUT_TENSOR_NAME = 'ImageTensor:0', OUTPUT_TENSOR_NAME = 'SemanticPredictions:0'):
        """Roda a inferência em uma imagem. 

        Parâmetros: 
            image: Objeto PIL.Image que contém a imagem de input. 
            INPUT_TENSOR_NAME: O nome do tensor de entrada. padrão=ImageTensor
            OUTPUT_TENSOR_NAME: O name do tensor de saída. padrão=SemanticPredictions

        Returns:
            resized_image: imagem de entrada RGB redimensionada  
            seg_map: Mapa de segmentação do `resized_image` 
        """
        width, height = image.size
        target_size = (2049,1025)  # tamanho das imagens do dataset Cityscapes
        resized_image = image.convert('RGB').resize(target_size, Image.ANTIALIAS)

        batch_seg_map = self.sess.run(
            OUTPUT_TENSOR_NAME,
            feed_dict={INPUT_TENSOR_NAME: [np.asarray(resized_image)]})
        
        seg_map = batch_seg_map[0]  # espera o batch size = 1
        if len(seg_map.shape) == 2:
            seg_map = np.expand_dims(seg_map,-1)  # adiciona uma dimensão extra, necessária pro cv2.resize

        seg_map = cv2.resize(seg_map, (width,height), interpolation=cv2.INTER_NEAREST)
        return seg_map

TARBALL_NAME = 'deeplab_model.tar.gz'
model_dir = tempfile.mkdtemp()
tf.io.gfile.makedirs(model_dir)
MODEL_NAME = 'mobilenetv2_coco_cityscapes_trainfine'
#MODEL_NAME = 'xception65_cityscapes_trainfine'
DOWNLOAD_URL_PREFIX = 'http://download.tensorflow.org/models/'
MODEL_URLS = {
    'mobilenetv2_coco_cityscapes_trainfine' : 'deeplabv3_mnv2_cityscapes_train_2018_02_05.tar.gz',
    'xception65_cityscapes_trainfine'       : 'deeplabv3_cityscapes_train_2018_02_06.tar.gz',
}
download_path = os.path.join(model_dir, TARBALL_NAME)
download_path
urllib.request.urlretrieve(DOWNLOAD_URL_PREFIX + MODEL_URLS[MODEL_NAME], download_path)
```

```python
# Parametros
kernel_size = 5
frame_inicio = 70  # Frame de início desejado
frame_fim = 71  # Frame de fim desejado
tamanho_fonte = 0.02  # 2% em relação à largura do frame
diamentro_da_centroide = 0.8 # tamnahdo da centroide  
tamanho_cruz = 5  # tamnahdo da cruz
posição_Y = 0.09  # 9% em relação ao topo do frame (distância da borda)
posicao_linha_esquerda = 0.50  # 80% em relação ao eixo x (da esquerda para direita)
percent_reducao = 0.5  # Percentual para redução da imagem para processamento mais rápido (100% equivale a 1, 0.6 equivale a 60% de redução)


# Variáveis
qtd_pessoas_id = 0
m_obj_id = 0
m_qtd_pessoas_id = 0


# Carregar o vídeo
cap = cv2.VideoCapture('/content/people6.mp4')
total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
arrayw = []
arrayh = []
frames = []
image_arrays = []
offset =10
frame_count = 0 
MIN_DISTANCE = 50   # pixels distancia minima para calcular se passou pela linha 
DETECTION_DELAY = 1.0   ## DETECTION_DELAY para calculo frame nao fircar mais acelerado que o fps normal

subtracao = cv2.createBackgroundSubtractorMOG2()
frame_fim = min(frame_fim, total_frames)

# Loop principal iniciando o lop para percorrer todos os frames  e fazer a segmentação
while True:
    ret, frame = cap.read()
    if not ret:
        break
    frameOriginal=frame

# Verificar se o frame está dentro do intervalo desejado colocado la em cima do  nos parametros
    if frame_count < frame_inicio:
        frame_count += 1
        continue
    elif frame_count >= frame_fim:
        break

# Redimensionar o frame
    frame = cv2.resize(frame, None, fx=percent_reducao, fy=percent_reducao)
    pil_image = Image.fromarray(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
    seg_map = model.run(pil_image)
    seg_map = seg_map.astype('uint8')  # Redimensionar arredondat  valor do array do frame para numeros inteiros.
    linha_esquerda = int(frame.shape[1] *posicao_linha_esquerda)  #calcula posicao da linha  da tela calculo em ralação ao tamanho da imagem para nao ficar muito pequeno nem muito grande a espes da linha
    tamanho_ponto =int(frame.shape[1] * 0.01 ) * diamentro_da_centroide  #calcula  o fator de escala conforme desejadoparmetro definido la em cima 
    tamanho_cruzes = int(tamanho_ponto * tamanho_cruz)    #calcula  o fator de escala conforme desejadoparmetro definido la em cima 
    global trackers   #criando um lista ddicionado ´para usar no objeto passou a linha 
    trackers = {}  
    global obj_id
    objetos_detectados = set()  #array que guarda objetos indetificados.

#---------------------------------------------------------------------------------------------------------------------------------------------------------------------
#----------------------------------------------SEGMENTAÇÃO UTILZIANDO INTELIGENCIA ARTIFICIAL-------------------------------------------------------------------------
#---------------------------------------------------------------------------------------------------------------------------------------------------------------------

# Função para visualizar a segmentação
#A função vis_segmentacaoia(image, seg_map) utiliza um modelo de segmentação pré-treinado INTELIGENCIA ARTIFICIALÇ para detectar pessoas na imagem de segmentação.
#Ela realiza a segmentação, identifica os contornos das pessoas, realiza o rastreamento das pessoas, desenha retângulos, linhas e cruzes 
# para visualização, conta a quantidade de pessoas detectadas e exibe informações na imagem. Por fim, cria uma imagem sobreposta combinando a imagem original
# e a imagem de segmentação com as detecções e retorna essa imagem para uso posterior.

    def vis_segmentacaoia(image, seg_map):
        global qtd_pessoas_id, trackers, obj_id, detections, objetos_detectados
        #Essa linha de código converte o mapa de segmentação em uma imagem colorida, destacando cada classe com uma cor específica. Em seguida, a imagem é convertida para o tipo de dados uint8.
        iaseg_image = label_to_color_image(seg_map).astype(np.uint8) 
        #Aqui criamos  uma máscara que identifica os pixels da imagem de segmentação que correspondem à classe das pessoas, classe 11 e pessoas faço isso para poder criar o retangulo em volta da pesssoa, e seguir na detectação.
        ia_person_mask = (seg_map == 11)
        ia_person_image = np.zeros_like(seg_map)
        #atribui o valor 255 (cor branca) para objeto pessoa o restante cor e preta.
        ia_person_image[ia_person_mask] = 255
        # fiz um tenste aplicando um a função cv2.threshold()
        #ela  usada para separar os pixels de uma imagem em preto e branco, utilizando um limiar de 127.
        # Pixels com valores acima de 127 são definidos como branco (255), enquanto pixels com valores abaixo de 127 são definidos como preto (0). 
        #Essa operação é útil para destacar áreas de interesse em uma imagem com base na intensidade de cor ou nível de cinza.
        _, ia_thresholded_image = cv2.threshold(ia_person_image, 127, 255, cv2.THRESH_TOZERO)
        #A função cv2.findContours() é usada para encontrar as bordas dos objetos em uma imagem. No código, ela é utilizada para encontrar os contornos das regiões na imagem binária ia_thresholded_image, após a operação de thresholding.
        #Depois de encontrar os contornos, podemos desenhar retângulos ao redor dessas regiões. Esses retângulos ajudam a destacar e delimitar as áreas de interesse na imagem.
        ia_contours, _ = cv2.findContours(ia_thresholded_image, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
        contornos_e_areas = []
        #Resumidamente, esse loop percorre os contornos encontrados na imagem  e seleciona aqueles que representam pessoas com base em critérios de área e proporção.
        #Para cada pessoa encontrada, um tracker é criado e adicionado a um dicionário para rastreamento posterior.
        for ia_contour in ia_contours:
            ia_x, ia_y, ia_w, ia_h = cv2.boundingRect(ia_contour)
            ia_area = ia_w * ia_h 
            contornos_e_areas.append((ia_contour, ia_area, ia_w, ia_h))

        contornos_e_areas.sort(key=lambda x: x[1], reverse=True)
        max_area = contornos_e_areas[0][1]


        #verifica-se se o contorno tem uma área maior ou igual a 50% da maior área encontrada 
        #e se a altura é maior que 1,3 vezes a largura. Essas condições são usadas para filtrar contornos que representam pessoas com base em seu tamanho e proporção
        # evitando criar retangulos em lugares que nao sao pessoas.
        #uma pessoa geralmente nao e quadrada 

        for ia_contour, ia_area, ia_w, ia_h in contornos_e_areas:
            if ia_area >= max_area * 0.5 and ia_h > ia_w * 1.3:  
                ia_x, ia_y, ia_w, ia_h = cv2.boundingRect(ia_contour)
                tracker = cv2.TrackerCSRT_create()  
                tracker.init(image, (ia_x, ia_y, ia_w, ia_h))
                trackers[obj_id] = (tracker, False)
                obj_id += 1

       #atualizo a posicçao da pessoa 
        detections = []
        for id, (tracker, cruzou_linha) in trackers.items():
            success, box = tracker.update(image)

            #si a atualização do tracker for bem-suce a posição da pessoa detectada e obtida a partir do retângulo delimitador e são 
            #e crio um retangulo em volta da pessoa . Em seguida, é verificado se a pessoa cruzou uma linha de detecção e se não foi 
            #detectada recentemente . Se atender a essas condições, atualizo a variavel contagem de pessoas.
            if success:
                ia_x, ia_y, ia_w, ia_h = map(int, box)
                cv2.rectangle(iaseg_image, (ia_x, ia_y), (ia_x + ia_w, ia_y + ia_h), (0, 255,0), int(tamanho_ponto)-1)
                centro_x = ia_x + ia_w // 2
                centro_y = ia_y + ia_h // 2
               
                cv2.line(frame, (centro_x - tamanho_cruzes, centro_y), (centro_x + tamanho_cruzes, centro_y), (0, 255, 0), 1)
                cv2.line(frame, (centro_x, centro_y - tamanho_cruzes), (centro_x, centro_y + tamanho_cruzes), (0, 255, 0), 1)
                cv2.circle(iaseg_image, (centro_x, centro_y), int(tamanho_ponto), (0, 255, 0), -1)
                cv2.line(iaseg_image, (linha_esquerda, 0), (linha_esquerda, iaseg_image.shape[0]), (0, 255, 0),int(tamanho_ponto)-1 )

                if not cruzou_linha and abs(centro_x - linha_esquerda) <= offset:
                    current_time = time.time()
                    recent_detection = next((d for d in detections if d['time'] > current_time - DETECTION_DELAY and
                    abs(d['position'][0] - centro_x) < MIN_DISTANCE and abs(d['position'][1] - centro_y) < MIN_DISTANCE), None)

                    if not recent_detection:
                        
                        if id not in objetos_detectados:
                            
                            objetos_detectados.add(id) 
                            
                        detections.append({'position': (centro_x, centro_y), 'time': current_time})
                        trackers[id] = (tracker, True)
                        qtd_pessoas_id += 1

        font = cv2.FONT_HERSHEY_SIMPLEX
        text = "Quantidade de pessoas IA = {}".format(qtd_pessoas_id)
        
        #desenhos na imagem para facilitar visualização
        fontScale = iaseg_image.shape[1] * tamanho_fonte / cv2.getTextSize("X", font, 1, 2)[0][0]
        textsize = cv2.getTextSize(text, font, fontScale, 2)[0]
        textX = (iaseg_image.shape[1] - textsize[0]) // 2
        textY = int(iaseg_image.shape[0] * posição_Y )
        cv2.putText(iaseg_image, text, (textX, textY), font, fontScale, (255, 255, 255), 2, cv2.LINE_AA)
        frame_text = "Frames = {}/{}".format(frame_count+1, frame_fim)
        frame_textsize = cv2.getTextSize(frame_text, font, fontScale, 2)[0]
        frame_textX = iaseg_image.shape[1] - frame_textsize[0] - 20  
        frame_textY = iaseg_image.shape[0] - frame_textsize[1] - 20  
        cv2.putText(iaseg_image, frame_text, (frame_textX, frame_textY), font, fontScale, (255, 255, 255), 2, cv2.LINE_AA)

        resized_frame = cv2.resize(image, (iaseg_image.shape[1], iaseg_image.shape[0]))
        overlay = cv2.addWeighted(resized_frame, 0.3, iaseg_image, 0.7, 0)

        
        return overlay
    imagem_sobreposta = vis_segmentacaoia(frame, seg_map)

   
#---------------------------------------------------------------------------------------------------------------------------------------------------------------------
#----------------------------------------------SEGMENTAÇÃO MANUAL CONFORME APRENDIDO EM SALA -------------------------------------------------------------------------
#---------------------------------------------------------------------------------------------------------------------------------------------------------------------
    #com  imagem ja em 
    
    #aplicO a subtração de fundo para obter uma máscara nao funciona para video one funco se move
    fg_mask = subtracao.apply(frame)
    #com  imagem ja em tons de cinza
    gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    kernel = np.ones((kernel_size, kernel_size), np.uint8)
    #utilizo  operação de abertura morfológica  para tentar  remover ruídos e detalhes indesejados de uma imagem . Ela consiste em duas etapas: erosão seguida de dilatação.
    opening = cv2.morphologyEx(fg_mask, cv2.MORPH_OPEN, kernel)
    #tento fazer os fechamento dos buracos aberto na parte de cima
    closing = cv2.morphologyEx(opening, cv2.MORPH_CLOSE, kernel)
    #crio 2 x a  dilatação das linhas para completar o fechamento das linhas 
    dilat = cv2.dilate(closing, np.ones((10, 10)))
    dilatacao = cv2.dilate(dilat, np.ones((3, 3), np.uint8))
    #Resumidamente, esse loop percorre os contornos encontrados na imagem  e seleciona aqueles que representam pessoas com base em critérios de área e proporção.
    #Para cada pessoa encontrada, um tracker é criado e adicionado a um dicionário para rastreamento posterior.
    contours, hierarchy = cv2.findContours(dilatacao, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
     # fiz um tenste aplicando um a função cv2.threshold()
    #ela  usada para separar os pixels de uma imagem em preto e branco, utilizando um limiar de 127.
    # Pixels com valores acima de 127 são definidos como branco (255), enquanto pixels com valores abaixo de 127 são definidos como preto (0). 
    #Essa operação é útil para destacar áreas de interesse em uma imagem com base na intensidade de cor ou nível de cinza.
    _, thresholded_frame = cv2.threshold(gray_frame, 127, 255, cv2.THRESH_TOZERO)
    _, inverted_frame = cv2.threshold(gray_frame, 127, 255, cv2.THRESH_TOZERO_INV)

        
    areas = []  
    maior_area = 0  
    #identifica contornos e as areas desse contornos
    for contour in contours:
        x, y, w, h = cv2.boundingRect(contour)
        area = w * h
        areas.append(area)
        for area in areas:
            if area > maior_area:
                maior_area = area
   #verifica-se se o contorno tem uma área maior ou igual a 50% da maior área encontrada 
   #e se a altura é maior que 1,3 vezes a largura. Essas condições são usadas para filtrar contornos que representam pessoas com base em seu tamanho e proporção
   # evitando criar retangulos em lugares que nao sao pessoas.
   #uma pessoa geralmente nao e quadrada                 
    for contour, area in zip(contours, areas):
        x, y, w, h = cv2.boundingRect(contour)
        if h >= w * 1.3 and area >= maior_area * 0.5:
            
            #atualizado meu dicionario
            tracker = cv2.TrackerCSRT_create()  
            tracker.init(frame, (x, y, w, h))
            trackers[m_obj_id] = (tracker, False)
            m_obj_id += 1

    #A função conta_pessoas() recebe uma lista de detecções e conta o número de pessoas que cruzaram uma linha específica na imagem. Cada detecção é processada e verificado se cruzou a linha.
    #Se cruzar, o contador é incrementado, uma linha é desenhada na imagem 
    #. As detecções que não cruzaram a linha são armazenadas em uma nova lista que é retornada pela função. 
    def conta_pessoas(deteccoes):
        global m_qtd_pessoas_id
        novas_deteccoes = []
        for deteccao in deteccoes:
            x, y = deteccao['x'], deteccao['y']
            if abs(x - linha_esquerda) <= offset:
                m_qtd_pessoas_id += 1
                cv2.line(frame, (linha_esquerda, 0), (linha_esquerda, frame.shape[0]), (0, 127, 255), 3)
                #print("Pessoas detectadas até o momento: " + str(m_qtd_pessoas_id))
            else:
                novas_deteccoes.append(deteccao)
        return novas_deteccoes

    m_detections = []
    for id, (tracker, cruzou_linha) in trackers.items():
        success, box = tracker.update(frame)
        if success:
            x, y, w, h = map(int, box)
            #desenhos
            cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0),int(tamanho_ponto)-1)
            centro_x = x + w // 2 #econtro centros dos retangulos
            centro_y = y + h // 2
            cv2.circle(frame, (centro_x, centro_y), int(tamanho_ponto), (0, 0, 255), -1)
            tamanho_cruzes = int(tamanho_ponto * tamanho_cruz)             
            cv2.line(frame, (centro_x - tamanho_cruzes, centro_y), (centro_x + tamanho_cruzes, centro_y), (0, 0, 255), 1)
            cv2.line(frame, (centro_x, centro_y - tamanho_cruzes), (centro_x, centro_y + tamanho_cruzes), (0, 0, 255), 1)
            cv2.line(frame, (linha_esquerda, 0), (linha_esquerda, frame.shape[0]), (255, 0, 0), int(tamanho_ponto)-1) 
            #codigo para garantir que pessoas sejavam contadas mais de uma ves uma ves que o frame atualize com centro em cima da linha 
            if not cruzou_linha and abs(centro_x - linha_esquerda) <= offset:
                current_time = time.time()
                recent_detection = next((d for d in m_detections if d['time'] > current_time - DETECTION_DELAY and abs(d['x'] - centro_x) < MIN_DISTANCE and abs(d['y'] - centro_y) < MIN_DISTANCE), None)
                if not recent_detection:
                    m_detections.append({'x': centro_x, 'y': centro_y, 'time': current_time})
                trackers[id] = (tracker, True)
    m_detections = conta_pessoas(m_detections)
    #desenhando  na imagem 
    font = cv2.FONT_HERSHEY_SIMPLEX
    text = "Quantidade de pessoas = {}".format(m_qtd_pessoas_id)
    fontScale = frame.shape[1] * tamanho_fonte / cv2.getTextSize("X", font, 1, 2)[0][0]
    textsize = cv2.getTextSize(text, font, fontScale, 2)[0]
    textX = (frame.shape[1] - textsize[0]) // 2
    textY = int(frame.shape[0] * posição_Y )
    frame_text = "Frames = {}/{}".format(frame_count+1, frame_fim)
    frame_textsize = cv2.getTextSize(frame_text, font, fontScale, 2)[0]
    frame_textX = frame.shape[1] - frame_textsize[0] - 20  
    frame_textY = frame.shape[0] - frame_textsize[1] - 20  
    cv2.putText(frame, frame_text, (frame_textX, frame_textY), font, fontScale, (255, 255, 255), 2, cv2.LINE_AA)
    cv2.putText(frame, text, (textX, textY), font, fontScale, (255, 255, 255), 2, cv2.LINE_AA)
    frames.append(frame)

    ret, _ = cap.read()
    if ret:
        clear_output(wait=True)  
    
    frameOriginalredimensionado = cv2.resize(frameOriginal, None, fx=percent_reducao, fy=percent_reducao) #ajusto tamanho do frame original para nao dar pau na insert do array

    
    #funcao para mostras todas imagem segmentadas juntas 2 primeira linhas sao iamgem segmentadas manualmente 
    #ultima  linha imagem segmentada pela INTELIGENCIA ARTIFICAL, A IDEIA SERIA FAZER UMA Comparação qual e melhor.
    def vis_segmentacao(image, seg_map):

   
        seg_image = label_to_color_image(seg_map).astype(np.uint8)
        fig = plt.figure(figsize=(30, 20))
        gs = GridSpec(4, 3, figure=fig)
        axes = [fig.add_subplot(gs[i//3, i%3]) for i in range(9)]
        ax10 = fig.add_subplot(gs[3, :]) 
        images = [cv2.cvtColor(frameOriginalredimensionado, cv2.COLOR_BGR2RGB),cv2.cvtColor(gray_frame, cv2.COLOR_GRAY2RGB)  , cv2.cvtColor(thresholded_frame, cv2.COLOR_GRAY2RGB)  ,cv2.cvtColor(opening, cv2.COLOR_GRAY2RGB) , cv2.cvtColor(dilat, cv2.COLOR_GRAY2RGB) , 
              cv2.cvtColor(image, cv2.COLOR_BGR2RGB) , cv2.cvtColor(frameOriginalredimensionado, cv2.COLOR_BGR2RGB), seg_image, imagem_sobreposta ]

        titles = ['Imagem Original', 'Imagem gray_frame', 'Imagem thresholded_frame',
                  'Imagem opening', 'Imagem dilatacao', 'Imagem Segmentacao Manual',
                  'Original', 'Mapa de segmentação', 'Sobreposição da segmentação']



        for ax, img, title in zip(axes, images, titles):
                if title == 'Imagem thresholded_frame':
                    ax.imshow(img, cmap='gray')  
                else:
                    ax.imshow(img)
                ax.axis('off')
                ax.set_title(title)
            
                unique_labels = np.unique(seg_map)
                ax10.imshow(FULL_COLOR_MAP[unique_labels].astype(np.uint8), interpolation='nearest')
                ax10.set_yticks(range(len(unique_labels)))
                ax10.set_yticklabels(LABEL_NAMES[unique_labels])
                ax10.set_xticks([])
                ax10.grid(False)
                ax10.set_title('Mapa de cores')
            
                
                image_arrays.append(np.array(ax.images[0].get_array()))
      
        plt.tight_layout()
        plt.show()
        return image_arrays  

    image_arrays = vis_segmentacao(frame, seg_map)

    frames.append(frame)
    frame_count += 1  


fig, axs = plt.subplots(len(image_arrays), figsize=(40, len(image_arrays)))


#adiciono todas imagens em um array para criar um video depois.
print ('terminou ')
for ax, img in zip(axs, image_arrays):
    ax.imshow(img)
    ax.axis('off')
    print ('passou pelo for  ')

plt.tight_layout()
plt.show()

   
   
cap.release()
```
